```python
import utilities as utils
import time
import numpy as np
import random as r
```

```python
# Constant
CHOICES = ['b', 's', 'n']
FEES = 0.1/100
FEES_BUY = 0.06/100
FEES_SELL = 0.02/100
```

```python
class Simulation:
    def __init__(self, matrix_coef, matrix, initialM):
        self.matrix_coef = matrix_coef
        self.matrix = matrix
        self.money = initialM
        self.investMoney = 0
        self.nb_buy = 0
        self.nb_sell = 0
        
    def doAStep(self, step):
        choice = r.randint(0, 200)
        if choice == 0:
            amoutToBuy = self.money
            self.buy(amoutToBuy)
        if choice == 1:
            self.sell()
        self.update(step)
    
    def buy(self, amout):
        self.nb_buy +=1
        self.money -= amout
        self.investMoney += (amout-amout*FEES_BUY)
    
    def sell(self):
        self.nb_sell +=1
        self.money += (self.investMoney-self.investMoney*FEES_SELL)
        self.investMoney = 0
    
    def update(self, step):
        priceEvolution = step[4]
        self.investMoney *= priceEvolution
    
    def getInfo(self, printing):
        if printing:
            print(f"Money : {self.money}$")
            print(f"Invest Money : {self.investMoney}$")
            print(f"Somme : {self.money+self.investMoney}")
            print(f"buy/sell : {self.nb_buy}/{self.nb_sell}")
        return [self.money, self.investMoney , self.money+self.investMoney, self.nb_buy , self.nb_sell]
    
    
```

```python
matrix_coef = utils.read_csv_to_numpy("binance_coef_data_of_ETHUSDT_15m_0_None.csv")
matrix = utils.read_csv_to_numpy("binance_data_of_ETHUSDT_15m_0_None.csv")
```

```python
print(f"\nSimulation start \n")
simulation = Simulation(matrix_coef, matrix, 1000)
for i, step in enumerate(matrix_coef):
    simulation.doAStep(step)

simulation.getInfo(True) 
print(f"\nSimulation end \n")
    
```

```python
all_info = []
for k in range(500):
    print(f"\nSimulation{k} \n")
    simulation = Simulation(matrix_coef, matrix, 1000)
    for i, step in enumerate(matrix_coef):
        simulation.doAStep(step)

    all_info.append(simulation.getInfo(False))

all_info = np.array(all_info)
print("Money : ", np.median(all_info[:, 2]))
print("nb buy : ", np.median(all_info[:, 3]))
print("nb sell : ", np.median(all_info[:, 4]))
    
    
```
